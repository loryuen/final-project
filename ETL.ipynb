```python
import pandas as pd
import numpy as np
```

```python
county_mobility = pd.read_csv("County_Mobility_Report.csv")
```

```python
Suffix_List = ["County"]
county_mobility["COUNTY"] = county_mobility["sub_region_2"].str.replace("\s?(" + "|".join(Suffix_List) + ")$", "")
county_mobility.head()
```

```python
combined_data = pd.read_csv("combined_csv.csv")
combined_data.head()
```

```python
grouped = pd.DataFrame(combined_data.groupby(['date','STATE','COUNTY']).mean()['DAILY_AQI_VALUE'])
grouped.head()
```
