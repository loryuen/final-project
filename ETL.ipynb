```python
import pandas as pd
import numpy as np
```

```python
county_mobility = pd.read_csv("data/County_Mobility_Report.csv")
```

```python
county_mobility.head()
```

```python
Suffix_List = ["County"]
county_mobility["COUNTY"] = county_mobility["COUNTY"].str.replace("\s?(" + "|".join(Suffix_List) + ")$", "")
county_mobility.head()
```

```python
AQI_changes = pd.read_csv("data/AQI_changes.csv")
AQI_changes.head()
```

```python
grouped = pd.DataFrame(AQI_changes.groupby(['date','STATE','COUNTY']).mean()['DAILY_AQI_VALUE'])
grouped.head()
```

```python
#grouped_by_state = grouped.unstack(level = 0)
#grouped_by_state.columns = grouped_by_state.columns.get_level_values(1)
#grouped_by_state
```

```python
#all_data = pd.merge(combined_data, county_mobility, how='outer', left_on=["STATE","COUNTY","date"], right_on=["STATE","COUNTY","date"])
#all_data = grouped_by_state.merge(county_mobility, how='left', left_on=["STATE","COUNTY"], right_on=["STATE","COUNTY"])
all_data = AQI_changes.merge(county_mobility, how='outer', left_on=["date","STATE","COUNTY"], right_on=["date","STATE","COUNTY"])
#merged_df = left_df.merge(right_df, how='inner', left_on=["A", "B"], right_on=["A2","B2"])
all_data.tail()
```
