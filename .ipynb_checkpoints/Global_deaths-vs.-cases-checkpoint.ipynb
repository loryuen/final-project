```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import timedelta
```

```python
# Read the csv file into a pandas DataFrame

df = pd.read_csv('covid-19-total-confirmed-cases-vs-total-confirmed-deaths.csv')
df.head()
```

```python
pd.options.display.float_format = '{:,.0f}'.format
df
```

```python
#Rename columns for readability
df_renamed = df.rename(columns={"Entity":"Continent",
                                                "Total confirmed deaths due to COVID-19 (deaths)": "Deaths",
                                                "Total confirmed cases of COVID-19 (cases)": "Cases"
                                                })
df_renamed.head() 
for col in df_renamed: 
    print (col)
```

```python
df_renamed.set_index('Continent',inplace=True)
df_renamed.head()
```

```python
df_renamed['Date'] = pd.to_datetime(df_renamed.Date)
print (df_renamed)

df_renamed['New Date'] = df_renamed['Date'].dt.strftime('%m/%d/%Y')
print (df_renamed)
```

```python
del df_renamed["Date"]
df_renamed.head()
```

```python
df_renamed.dtypes 
```

```python
df_renamed2 = df_renamed.rename(columns={"New Date":"Date"
                                            
                                                })
df_renamed2.head()
for col in df_renamed2: 
    print (col)
```

```python
# Obtain the x and y coordinates for each of the three city types
Asia = df_renamed2[df_renamed2["Continent"] == "Asia"]
Australia = df_renamed2[df_renamed2["Continent"] == "Australia"]
North_Amer = df_renamed2[df_renamed2["Continent"] == "North America"]
South_Amer = df_renamed2[df_renamed2["Continent"] == "South America"]
Africa = df_renamed2[df_renamed2["Continent"] == "Africa"]
Europe = df_renamed2[df_renamed2["Continent"] == "Europe"]

Asia_count_days = Asia.groupby(["Continent"]).count()["Date"]
Asia_deaths = Asia.groupby(["Continent"]).count()["Deaths"]
Asia_cases = Asia.groupby(["Continent"]).mean()["Cases"]

Australia_count_days = Australia.groupby(["Continent"]).count()["Date"]
Australia_deaths = Australia.groupby(["Continent"]).count()["Deaths"]
Australia_cases = Australia.groupby(["Continent"]).mean()["Cases"]

Africa_count_days = Africa.groupby(["Continent"]).count()["Date"]
Africa_deaths = Africa.groupby(["Continent"]).count()["Deaths"]
Africa_cases = Africa.groupby(["Continent"]).mean()["Cases"]

NorthAmer_count_days = North_Amer.groupby(["Continent"]).count()["Date"]
NorthAmer_deaths = North_Amer.groupby(["Continent"]).count()["Deaths"]
NorthAmer_cases = North_Amer.groupby(["Continent"]).mean()["Cases"]

SouthAmer_count_days = South_Amer.groupby(["Continent"]).count()["Date"]
SouthAmer_deaths = South_Amer.groupby(["Continent"]).count()["Deaths"]
SouthAmer_cases = South_Amer.groupby(["Continent"]).mean()["Cases"]

Europe_count_days = Europe.groupby(["Continent"]).count()["Date"]
Europe_deaths = Europe.groupby(["Continent"]).count()["Deaths"]
Europe_cases = Europe.groupby(["Continent"]).mean()["Cases"]
```

```python
# Build the scatter plots for each city types
plt.scatter(Asia_cases, 
            Asia_deaths, 
            s=10*Asia_count_days, c="pink", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Asia")

plt.scatter(Australia_cases, 
            Australia_deaths, 
             s=10*Australia_count_days,c="lightblue", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Australia")

plt.scatter(Africa_cases, 
            Africa_deaths, 
             s=10*Africa_count_days,c="gold", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Africa")

plt.scatter(NorthAmer_cases, 
            NorthAmer_deaths, 
            s=10*NorthAmer_count_days,c="lightGreen", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="North America")

plt.scatter(SouthAmer_cases, 
            SouthAmer_deaths, 
            s=10*SouthAmer_count_days,c="purple", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="South America")

plt.scatter(Europe_cases, 
            Europe_deaths, 
            s=10*Europe_count_days,c="brown", 
            edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Europe")


# Incorporate the other graph properties
plt.title("Global Death Vs. Cases from Covid-19 (Jan.2020-May.2020)")
plt.ylabel("Average Cases")
plt.xlabel("Total Deaths")
plt.grid(True)

# Create a legend
lgnd = plt.legend(fontsize="small", mode="Expanded", 
                  numpoints=1, scatterpoints=1, 
                  loc="best", title="Continents", 
                  labelspacing=0.5)
lgnd.legendHandles[0]._sizes = [50]
lgnd.legendHandles[1]._sizes = [50]
lgnd.legendHandles[2]._sizes = [50]
lgnd.legendHandles[3]._sizes = [50]
lgnd.legendHandles[4]._sizes = [50]
lgnd.legendHandles[5]._sizes = [50]



# Incorporate a text label regarding circle size
plt.text(50, 30,"Note:\nCircle size correlates with dates with confirmed cases and/or deaths.")

plt.savefig("/GlobelBubbleGraph.png")

# Show plot
plt.show()
```
