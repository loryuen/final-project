```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
covid=pd.read_csv("COVID-19-geographic-disbtribution-worldwide-2020-04-30.csv")
covid = covid.rename(columns = {"dateRep":"Date", "geoId":"Country"})
covid=covid.set_index('Date')
covid=covid.sort_index(ascending = True)
covid.head()
```

```python
air=pd.read_csv("../2020_combined.csv")
countries = ["US", "GB", "IT", "CN", "IN"]
air = air[air.Country.isin(countries)]
air=air.set_index('Date')
air=air.drop(columns = ['Unnamed: 0'])
air=air.sort_index(ascending = True)
air.to_csv('filtered_air.csv')
air.head()
```

```python
all= pd.merge(covid, air,  how='left', left_on=['Date','Country'], right_on = ['Date','Country'])
all= all.dropna(subset=['Specie'])
all.head()
```

```python
all["Death_rate"] = round((all['deaths'] / all['popData2018'])*100,2)
all["Case_rate"] = (all['cases'] / all['popData2018'])*100
all["death/case_rate"] = (all['deaths'] / all['cases'])*100
all = all.reset_index()
all.to_csv("all.csv")
all.head()
```
